```python
with open('./content_data.csv') as f:
    lines = f.read()
lines = lines[7:] # 앞에 content 
lines=" ".join(lines.split())
print(lines[0:100])
```

```python
lines = re.sub(r"[\[\"\']","",lines) # [, ', " 제거
lines = re.sub(r"[\\n]","",lines) # [, ', " 제거
lines = re.sub(r"(\] )", ".", lines) # '] '를 '.'으로 수정
print(lines[0:500])
```

```python
output = open("./output_data.txt", 'w')
output.write(lines)
output.close()
```
